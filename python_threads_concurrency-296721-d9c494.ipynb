```python
Конкурентное программирование — это область программирования, связанная с
одновременным выполнением нескольких задач или процессов. Вот основные термины,
которые помогут вам разобраться в этой теме:
1.	Процесс (Process): Независимая программа, выполняющаяся на компьютере. Процессы 
имеют свои собственные ресурсы (память, файловые дескрипторы и т.д.) и изолированы друг от друга.
2.	Поток (Thread): Легковесный процесс, который выполняется внутри процесса. Потоки 
могут совместно использовать ресурсы своего родительского процесса, такие как память и файловые дескрипторы.
3.	Многопоточность (Multithreading): Техника, позволяющая выполнять несколько потоков
одновременно внутри одного процесса. Это может улучшить производительность приложения за
счет лучшего использования ресурсов процессора.
4.	Многозадачность (Multitasking): Способность операционной системы выполнять несколько
задач одновременно. Это может быть достигнуто с помощью многопоточности или мультипроцессинга.
5.	Мультипроцессинг (Multiprocessing): Метод выполнения нескольких процессов одновременно.
Это более тяжелый по ресурсам метод по сравнению с многопоточностью, но он может обеспечить
лучшую изоляцию и стабильность.
6.	Конкурентность (Concurrency): Способность системы управлять несколькими задачами 
одновременно. Конкурентность не обязательно означает параллельное выполнение задач,
но гарантирует, что задачи будут прогрессировать.
7.	Параллелизм (Parallelism): Форма конкурентности, где задачи действительно выполняются
одновременно, например, на разных ядрах процессора.
8.	Синхронизация (Synchronization): Механизм управления доступом к общим ресурсам для 
предотвращения конфликтов. Примеры: блокировки (locks), семафоры (semaphores), барьеры (barriers).
9.	Блокировка (Lock): Механизм синхронизации, который ограничивает доступ к ресурсу 
до тех пор, пока он не будет освобожден.
10.	Семафор (Semaphore): Более гибкий механизм синхронизации, который управляет доступом
к нескольким копиям ресурса.
11.	Мьютекс (Mutex): Специфический тип блокировки, который используется для 
предотвращения одновременного доступа к ресурсу несколькими потоками.
12.	Гонка (Race Condition): Ситуация, когда результат программы зависит от порядка
выполнения потоков. Это обычно происходит при неправильной синхронизации потоков.
13.	Дедлок (Deadlock): Состояние, при котором два или более потоков застревают,
ожидая друг друга освободить ресурсы. Это приводит к бесконечному ожиданию и остановке выполнения.
14.	Живучесть (Liveness): Свойство системы, которое гарантирует, что все процессы 
или потоки продолжат свое выполнение и завершатся.
15.	Голодание (Starvation): Ситуация, когда поток или процесс не может получить доступ к 
необходимым ресурсам и, следовательно, не может прогрессировать.
16.	Барьер (Barrier): Механизм синхронизации, который останавливает выполнение потоков до 
тех пор, пока все они не достигнут определенной точки.
```

```python
CPython и GIL (Global Interpreter Lock)
1.	Global Interpreter Lock (GIL):
o	Что это: GIL — это глобальная блокировка интерпретатора, которая ограничивает выполнение 
только одного потока Python байт-кода в любой момент времени.
o	Почему он существует: GIL был введен для упрощения реализации интерпретатора и управления 
памятью. Он помогает избежать проблем с потокобезопасностью в самом интерпретаторе.
o	Последствия:
	Ограничение параллелизма: GIL значительно ограничивает возможности параллельного 
    выполнения Python кода, особенно в многопоточных CPU-интенсивных задачах.
	Польза в I/O задачах: В задачах, связанных с вводом-выводом (например, сетевые операции, файловые операции), 
    GIL менее критичен, так как потоки часто ожидают завершения операций, не потребляя много CPU.
2.	Пути обхода GIL:
o	Мультипроцессинг: Использование модуля multiprocessing для создания отдельных процессов,
    которые не разделяют GIL. Каждый процесс имеет свой собственный интерпретатор Python и память.
o	Расширения на C: Написание расширений на C, которые могут временно освобождать GIL во время 
    выполнения CPU-интенсивных задач.
o	Альтернативные реализации Python: Использование альтернативных реализаций Python, таких как
    Jython или IronPython, которые не имеют GIL, или PyPy, который имеет оптимизированный GIL.
```

```python
JVM и многопоточность
1.	Модель памяти JVM:
o	Потокобезопасность: JVM предоставляет более мощные механизмы для работы с потоками по сравнению
с CPython. Она поддерживает настоящую многопоточность без глобальной блокировки.
o	Синхронизация: Для обеспечения безопасности данных при доступе из нескольких потоков используется
встроенная поддержка синхронизации (ключевые слова synchronized, классы Lock, ReentrantLock и т.д.).
2.	Ограничения:
o	Контекстные переключения: Частые контекстные переключения между потоками могут снижать производительность.
o	Сложность разработки: Разработка многопоточных приложений на JVM требует понимания проблем 
синхронизации, взаимоблокировок и других типов конкуренции.
o	Память и ресурсы: Создание большого количества потоков требует значительных ресурсов памяти
и может привести к исчерпанию ресурсов.
```

```python
import threading
import time

def worker():
    for _ in range(5):
        print(threading.current_thread().name, "is working")
        time.sleep(1)
    threads = [threading.Thread(target=worker) for _ in range(3)]

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()

# without Gil
import multiprocessing
import time

def worker():
    for _ in range(5):
        print(multiprocessing.current_process().name, "is working")
        time.sleep(1)
processes = [multiprocessing.Process(target=worker) for _ in range(2)]

for process in processes:
    process.start()

for process in processes:
    process.join()
```

```python
Примитивы синхронизации в Python
1.	Lock (Блокировка):
o	Простой примитив синхронизации, позволяющий одному потоку захватить блокировку и блокировать
другие потоки до освобождения.
o	Модуль: threading
2.	RLock (Рекурсивная блокировка):
o	Позволяет одному и тому же потоку захватить блокировку несколько раз, не вызывая дедлок.
o	Модуль: threading
3.	Semaphore (Семафор):
o	Обеспечивает ограничение на количество потоков, которые могут одновременно захватить семафор.
o	Модуль: threading
4.	BoundedSemaphore (Ограниченный семафор):
o	Подкласс семафора, который не позволяет увеличивать счётчик выше начального значения.
o	Модуль: threading
5.	Event (Событие):
o	Используется для уведомления потоков о каком-либо событии. Потоки могут ожидать установки флага события.
o	Модуль: threading
6.	Condition (Условие):
o	Позволяет одному или нескольким потокам ожидать до тех пор, пока не будет выполнено определённое условие.
o	Модуль: threading
7.	Barrier (Барьер):
o	Примитив синхронизации, который позволяет группе потоков дождаться друг друга.
o	Модуль: threading
```

```python
Конкурентность (Concurrency)
Конкурентность — это способность системы обрабатывать несколько задач одновременно, но не 
обязательно выполнять их параллельно. В конкурентной системе задачи могут прогрессировать 
независимо друг от друга, даже если они не выполняются одновременно.
•	Пример: Однопоточное приложение, которое чередует выполнение разных задач, например, 
обработка ввода пользователя и обновление интерфейса.
•	Цель: Улучшение отзывчивости и эффективности использования ресурсов.
Параллелизм (Parallelism)
Параллелизм — это одновременное выполнение нескольких задач. Для параллельного выполнения 
требуется наличие нескольких процессоров или ядер.
•	Пример: Многопоточное приложение, где несколько потоков выполняются одновременно на разных ядрах процессора.
•	Цель: Ускорение выполнения задач за счет их одновременного выполнения.
IO-bound и CPU-bound задачи
IO-bound задачи
IO-bound задачи ограничены скоростью ввода/вывода (например, чтение/запись на диск, сетевые операции). 
Они проводят большую часть времени в ожидании завершения операций ввода-вывода.
•	Пример: Сетевые запросы, чтение и запись файлов.
•	Подход: Для улучшения производительности таких задач можно использовать асинхронное 
программирование или многопоточность, чтобы можно было обрабатывать другие задачи, пока идет ожидание
    завершения IO операций.
CPU-bound задачи
CPU-bound задачи ограничены скоростью процессора. Они требуют интенсивного использования процессорного 
    времени для выполнения вычислений.
•	Пример: Обработка больших массивов данных, сложные математические вычисления.
•	Подход: Для улучшения производительности таких задач можно использовать параллелизм, распределяя 
    вычисления между несколькими ядрами процессора.
 
```
